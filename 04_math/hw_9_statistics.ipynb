Создайте случайную величину из [любого выбранного вами распределения](https://docs.scipy.org/doc/scipy/reference/stats.html), для разных значений `n` сгенерируйте 1000 выборок размера `n`. Постройте гистрограммы средних этих выборок и сделайте выводы.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
rv = stats.erlang(1, loc=0, scale=2)
sample = rv.rvs(size=10000, random_state=42)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
def sample_hist(sample_size):
    sample_averages = np.array([])

    samples_count = 1000

    for number in range(samples_count): 
        sample = rv.rvs(size=sample_size)
        sample_averages = np.append(sample_averages, sample.mean())
        
    label = 'sample of ' + str(sample_size) + ' histogram'
    plt.hist(sample_averages, density=True, label=label)
    plt.legend() 
    plt.ylabel('samples means')
    plt.xlabel('$x$')
    plt.show()
```

```python
for size in np.arange(50, 1000, 50):
    sample_hist(size)
```

Благодаря полученным гистограммам видим подтверждение центральной предельной теоремы: средние по выборкам демонстрируют нормальное или приближенное к нормальному распределение.
