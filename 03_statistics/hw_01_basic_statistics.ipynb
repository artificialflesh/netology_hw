```python
# считаю, что категория 9 проставлена ошибочно вместо 2, меняем все на 2
horse_data['age'].replace(9, 2, inplace=True)
horse_data['age'].value_counts()
```

```python
# стоит проверить нет ли явно ошибочно проставленных категорий по другим столбцам с категориями

col_categories = ['surgery', 'temperature of extremities', 'pain', 'outcome']

for col in col_categories:
    print(f'{col} - {horse_data[col].unique()}')
    
# с остальными столбцами все в порядке
```

```python
# отделим строки, в которых NaN более чем в 3 столбцах
# слишком много неизвестных, при необходимости можем обработать отдельно
missing_data = horse_data.loc[ horse_data.isna().sum(axis=1) > 3 ]
```

```python
missing_data
```

```python
# вынесли 10 строк в отдельный фрейм, в результате оставляем 290, в которых и будем заполнять пропуски
horse_data = horse_data.loc[ ~(horse_data.isna().sum(axis=1) > 3) ]
horse_data
```

```python
# по столбцам с категориями выбросы искать нет смысла, проверим выбросы по остальным столбцам с выводом
# количества выборосов

def outliers(column):
    q1 = horse_data[column].quantile(0.25)
    q3 = horse_data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    check = horse_data[~horse_data[column].between(lower_bound, upper_bound, inclusive='both')]
    count = check[column].count()
    min_ = check[column].min()
    max_ = check[column].max()
    return f'{column} outliers - {count}, min - {min_}, max - {max_}'
```

```python
col_numeric = ['rectal temperature', 'pulse', 'respiratory rate']
```

```python
# выбросы не содержат аномальных значений, оставляем их в датафрейме

list(map(outliers, col_numeric))
```

```python
# разделим датафрейм по возрастным категориям 1 и 2 для заполнения пропусков
horse_data_age_1 = horse_data.loc[ horse_data['age'] == 1 ]
horse_data_age_2 = horse_data.loc[ horse_data['age'] == 2 ]
```

```python
# заполним столбцы с категориями модой, столбцы с числовыми значениями медианой
def nan_filler(dataframe):
    dataframe['surgery'].fillna(dataframe['surgery'].mode()[0], inplace=True)
    dataframe['temperature of extremities'].fillna(dataframe['temperature of extremities'].mode()[0], inplace=True)
    dataframe['pain'].fillna(dataframe['pain'].mode()[0], inplace=True)
    dataframe['outcome'].fillna(dataframe['outcome'].mode()[0], inplace=True)
    
    dataframe['rectal temperature'].fillna(dataframe['rectal temperature'].median(), inplace=True)
    dataframe['pulse'].fillna(dataframe['pulse'].median(), inplace=True)
    dataframe['respiratory rate'].fillna(dataframe['respiratory rate'].median(), inplace=True)
```

```python
nan_filler(horse_data_age_1)
nan_filler(horse_data_age_2)
```

```python
# объединим датафреймы

horse_data = pd.concat([horse_data_age_1, horse_data_age_2])
horse_data.reset_index(drop=True)
```
